```
# Your code here
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
cereal = pd.read_csv('/content/drive/MyDrive/Data/cereal.csv')
cereal.head(2)
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
cereal.dropna()
cereal.count()
cereal.info()
```

```
cereal.rename(columns={'name': 'Brand', 'sugars': 'Sugar(G)', 'rating': 'Rating'}, inplace=True)
sugar_rating = cereal.groupby(by=['Brand', 'Rating','Sugar(G)'], as_index=False).count()
#sugar_rating.sort_values(by=['Sugar(G)','Rating'])
sns.relplot(kind='scatter', data=sugar_rating, x='Rating', y='Sugar(G)', aspect=2, )
```

```
sugar_rating.sort_values(by='Rating').tail(1)
```

```
chart = sns.relplot(
   kind='scatter',
   data=sugar_rating,
   x='Rating',
   y='Sugar(G)',
   aspect=2,
   marker='o',
   hue='Sugar(G)',
   s=200
)


ax = chart.axes[0,0]
#ax.text(x=17, y=15, s='Most Sugar', size='large', color='purple', weight='semibold')
#ax.text(x=70, y=2, s='Zero Sugar', size='large', color='#dd6061', weight='semibold')
ax.text(x=87, y=.75, s='All-Bran with Extra Fiber', size='small', color='pink', weight='semibold')
ax.text(x=15.5, y=12.75, s="Cap'n'Cr.", size='small', color='purple', weight='semibold')

ax.set(title="Cereal: Sugar And It's Relationship To Rating")
plt.show()
```

With my visualization I set out to understand the relationship between the amount of sugar in each cereal brand in the data set and it's rating. I anticipated the data would show a positive relationship between sugar and rating but was surprised by a negative corelation. To highlight this for the viewer I added the name of the brand with the highest rating (All-Bran), the lowest rating (Cap'n'), and used hue to show the change in distribution as sugar decreases.
